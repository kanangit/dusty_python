```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
outfile = r'G:\My Drive\workAppState\prj_shocks\expData\data_phonon_spectrum\data_set1\analyzed\140Vbias\20230608phonon\07b_Originfrequencies\outputs\freqs_trans.png'
```

```python
data = np.loadtxt(r'G:\My Drive\workAppState\prj_shocks\expData\data_phonon_spectrum\data_set1\analyzed\140Vbias\20230608phonon\07b_Originfrequencies\inputs\phon_Trans.csv', delimiter=',', dtype=float)
```

```python
data.shape
```

```python
arr_f = data[:,0]
arr_intens = data[:,1]
```

```python
len(arr_f)
```

```python
fff = np.sort(np.unique(arr_f))
```

```python
fff = np.delete(fff,0)
```

```python
arr_intens_counts = np.array([])
```

```python
len(arr_f)
```

```python
for elem in fff:
    arr_cur_f = arr_f[arr_f == elem]
    arr_cur_intens = arr_intens[arr_f == elem]
    cur_sum_intens = np.sum(arr_cur_intens)
    arr_intens_counts = np.append(arr_intens_counts, cur_sum_intens)
```

```python
fig, ax = plt.subplots()
ax.plot(fff, arr_intens_counts)
ax.set_xlabel('frequency (Hz)')
ax.set_ylabel('intensity (a.u.)')
ax.set_xlim([0, 150])
fig.savefig(outfile)
```
