```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

# Recall: Distribution Function and Probability Density Function

## Examples

## Distribution Function (Cumulative Distribution Function)

$$F(x) = P(X\le x)$$

## Probability Density Function

$$f(x):\text{ }F(x) = \int\limits_{- \infty}^{x}{f(u)du}$$
$$\int\limits_{a}^{b}{f(x)dx} = P(a \le X \le b)$$
$$\int\limits_{- \infty}^{+ \infty}{f(x)dx} = P(- \infty \le X \le  + \infty) = 1$$

# Uniform distribution

$$f(x) = \left\{ \begin{align}
  & \frac{1}{b-a},\text{ }x\in [a,b] \\ 
 & 0,\text{ }x\notin [a,b] \\ 
\end{align} \right.$$


```python
a = 1
b = 4

uniform_rv = sts.uniform(a, b-a)

uniform_rv.rvs(10)
```

```python
x = np.linspace(0,5,100)
cdf = uniform_rv.cdf(x)
fig, ax = plt.subplots()
ax.plot(x, cdf)
ax.set_ylabel('$F(x)$')
ax.set_xlabel('$x$')
```

Probability Density Function

```python
uniform_rv.pdf(3)
```

```python
x = np.linspace(0,5,1000)
pdf = uniform_rv.pdf(x)  # the function can receive a vector of numbers
fig, ax = plt.subplots()
ax.plot(x, pdf)

ax.set_ylabel('$f(x)$')
ax.set_xlabel('$x$')
fig.savefig('uniform.svg', format = 'svg')
```

# Normal Distribution (Gaussian)

$$ X \sim N(\mu, \sigma^2) $$
$$\mu \text{ characterizes mean, } \sigma^2 \text{ characterizes scatter}$$

$$f(x) = \frac{1}{\sqrt{2 \pi \sigma^2}}e^{-\frac{(x- \mu)^2}{2 \sigma^2}}$$

 $\mu=2.0$ and $\sigma=0.5$:

```python
mu = 2.0
sigma = 0.5

norm_rv = sts.norm(loc=mu, scale=sigma)

# generate 10 values
norm_rv.rvs(size=10)
```

### Cumulative Distribution Function

```python
norm_rv.cdf(3)
```

#### plot it:

```python
x = np.linspace(0,4,100)
cdf = norm_rv.cdf(x) # the function can receive a vector of numbers
fig, ax = plt.subplots()
ax.plot(x, cdf)
ax.set_ylabel('$F(x)$')
ax.set_xlabel('$x$')
```

### Probability Density Function

```python
norm_rv.pdf(3)
```

##### Plot it:

```python
x = np.linspace(0,4,100)
pdf = norm_rv.pdf(x)
fig, ax = plt.subplots()
ax.plot(x, pdf)

ax.set_ylabel('$f(x)$')
ax.set_xlabel('$x$')
```
